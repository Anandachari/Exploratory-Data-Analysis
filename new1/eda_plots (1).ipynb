```python
import pandas as pd

# Sample dataset
data = {
    "Name": ["Anand", "Priya", "Ravi", "Sneha", "Kiran", "Meena", "Arjun", "Lakshmi"],
    "Age": [22, 21, 23, 22, 24, 21, 23, 22],
    "Gender": ["Male", "Female", "Male", "Female", "Male", "Female", "Male", "Female"],
    "Course": ["Python", "Java", "Python", "Python", "Java", "C++", "Python", "Java"],
    "Marks": [85, 78, 90, 88, 76, 82, 95, 80]
}

df = pd.DataFrame(data)
print(df)
```

```python
# Check first rows
print(df.head())

# Dataset shape (rows, columns)
print(df.shape)

# Summary of dataset
print(df.info())

# Descriptive statistics
print(df.describe())

# Unique values in categorical columns
print(df["Course"].value_counts())
print(df["Gender"].value_counts())

# Check missing values
print(df.isnull().sum())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Histogram of Marks
plt.hist(df["Marks"], bins=5, edgecolor="black", color="skyblue")
plt.title("Distribution of Marks")
plt.xlabel("Marks")
plt.ylabel("Frequency")
plt.show()
```

```python
# 2. Bar plot: Count of Students per Course
sns.countplot(x="Course", data=df,color="yellow")
plt.title("Number of Students in Each Course")
plt.show()

```

```python
# 3. Boxplot: Marks by Gender
sns.boxplot(x="Gender", y="Marks", data=df)
plt.title("Marks Distribution by Gender")
plt.show()

```

```python
# 4. Scatter Plot: Age vs Marks
plt.scatter(df["Age"], df["Marks"], color="green")
plt.title("Age vs Marks")
plt.xlabel("Age")
plt.ylabel("Marks")
plt.show()
```
